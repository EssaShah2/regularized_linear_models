```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Ridge , RidgeCV
from sklearn.metrics import r2_score, root_mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ridge_regression_dataset.csv")
```

```python
x = df.drop(columns = "salary")
y = df["salary"]
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
df
```

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
```

```python
print(r2_score(y_test , y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

# Now we will use Ridge Regularized

```python
RG = Ridge(alpha = 5000) #by default alpha is 1 so i am using default value
RG.fit(x_train , y_train)
y_pred_RG = RG.predict(x_test)
```

```python
print(r2_score(y_test,y_pred_RG))
print(root_mean_squared_error(y_test,y_pred_RG))
```

# Ridge Cv

```python
alphas = [0.1,1,10,100,1000,5000,10000]
RGC = RidgeCV(alphas = alphas , cv = 10)
RGC.fit(x_train,y_train)
y_pred_RGC = RGC.predict(x_test)
```

```python
print(RGC.alpha_)
print(r2_score(y_test,y_pred_RGC))
```
